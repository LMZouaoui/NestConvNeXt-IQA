```python
import numpy as np
import torch
from predict import Net, predict_one_image

from PIL import Image
import torchvision.transforms as transforms


# path to model
svPath='./Pretrained models/model_LWIRIQA'  #####  model_LWIRIQA  ######## model_TIIQAD

# read image
img_path = "F:/LWIR_dataset/LWIR_IQA_dataset-main/ALL_IMAGES/img_db/IMG_01_J2K (5).bmp"

# number of patches in the test
test_patch_num = 10


transform = transforms.Compose([
					transforms.Grayscale(num_output_channels=1),				
					transforms.Resize((512, 512)),
					transforms.RandomCrop(size=224),
					transforms.ToTensor(),
					transforms.Normalize(mean=[0.5], std=[0.5])])


device = torch.device("cuda", index=int(0))
model = Net().to(device)

img = Image.open(img_path).convert("RGB")

pred_scores = np.zeros(test_patch_num, dtype=np.float64)
for ii in range(test_patch_num):
# convert to tensor
    image_tensor = transform(img)
    # example
    image_tensor = image_tensor.unsqueeze(0) 

    score = predict_one_image(
        Network=model,
        img=image_tensor,
        svPath=svPath,
        device=device,
        pretrained=1
    )
  
    pred_scores[ii] = score

predict_scores = np.mean(pred_scores)
print("Predicted score:", predict_scores)
```
